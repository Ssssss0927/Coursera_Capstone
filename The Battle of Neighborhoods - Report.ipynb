# Capstone Project - The Battle of Neighborhoods Report

## 1. Introduction

This project is to explore and make a comparison of different coffee shops in Tokyo, Japan and Toronto, Canada. Total coffee consumption in Japan has increased gradually over the last decade, reaching a consumption of over 7.8 million 60kg bags of coffee in 2018. The increasing trend is due to consumer preferences changed from tea to coffee. According to the Coffee Association of Canada, 2/3 of Canadians enjoy at least one cup a day with the average at 3.2 cups a day in 2018. Personally, I am a coffee lover, so I am interested in finding out what is the most popular coffee shop or brand in Tokyo and Toronto, whether it is a local brand or an imported brand. What role is Starbucks, as the largest coffee company in the world, playing in the coffee market in Tokyo and Toronto?

## 2. Data Section

Data used in this project is based on results from Foursquare. I find out the latitude and longitude of Tokyo and Toronto, then limited the radius to 500 to ensure managable data size. Then I used query endpoint to find out all the coffee shops in the limited area, including id, location, name, postal codes and other information. Some information is missing from the results from Foursquare, however, this did not impact on this project as the unique id is used as the priorty distinguisher. The next step is to filter and cleane the data for visulisation purposes (bar charts).

## 3. Methodology Section

### 3.1 Foursquare

The latitude and longitude information of Tokyo and Toronto, the selected locations for this project, is acquired through foursquare geolocator function. The next step is was to use the venue search end point of foursquare to find out all the coffee shops in the selected locations within the radius which was limited to 500 to ensure managable sample size. The results returned were in json format, which were then cleaned to a table containing two columns, name and id, through various fuctions in python, including filter, drop, rename, extract and so on.

### 3.2 Market Share Calculation

After data being cleaned, the tables only contain the information that is relevant to the purpose of the project, finding out the market share of different coffee brands in the selected locations. Then the percentage calculation function in python was used and the results were rounded to 2 deecimal places. The percentages gave a straightforward information.

### 3.3 Visualisation - Bar Chart

Matplotlib and matplotlib.pyplot were imported to python to help generate bar charts for the coffee shop market share in Tokyo and Toronto, which is a much easier way to deliver the conclusion. 

## 4. Results

In this project only, Tully's coffee took the largest market share (46.67%) in Tokyo, which was followed by Starbucks with 20.00%. The rest coffee shops included 5 Crossties Coffee, Doutor Coffee Shop, On-the-Rock CoffeeRoom and Seattle's Best Coffee, which were local coffee brands.

For Canada, 30 results were returned by foursquare. Starbucks took the largest market share in the radius set by this project, which was followed by Timonthy's World Coffee. 

## 5. Discussion

On comparison, Toronto had more coffee shops than Tokyo in the same radius and a lot more local coffee brands, which was not surprising due to Greater Toronto Area being 3.3 times as big as Tokyo. However, Starbucks took the largest market share in Toronto but only took. the second place in Tokyo. This could due to the fact that Starbucks menu prices don't include tax, therefore when applying tax the price was not competative to other brands, like Tully's and Doutor. Doutor is a very popular local coffee brand as it gives customers the most return for money. Tully's can provide you with somthing slightly fancy to go along with a small snack.

Tim Hortons, the largest coffee shop brand in Canada, only had 1 store, which could due to the radius set by the project. More research to cover larger areas should be conducted to get a more comprehensive conclusion.
